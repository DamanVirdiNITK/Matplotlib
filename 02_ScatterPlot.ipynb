```python
#plotting series of Points

x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

plt.scatter(x_values , y_values , s = 10)
plt.plot(x_values , y_values)
plt.show()

```

Calculating Data Automatically

```python

print(x_values)
y_values =[i**2 for i in x_values]
print(y_values)
#x**3 for x in x_values
```

```python
x_values = list(range(1 , 10))
print(x_values)

y_values = []
for i in x_values:
    y_values.append(i**2)
    
print(y_values)
```

```python
x_values = list(range(1 , 20))
print(x_values)
y_values = [x**3 for x in x_values]
sq_root = [x**(1/2) for x in x_values]
print(sq_root)

y_values2 = [x**2 for x in x_values]
print(y_values)
plt.scatter(x_values , y_values , s =100)
plt.scatter(x_values , y_values2 , s = 70)
plt.scatter(x_values , sq_root , c = 'red')
plt.show()
```

```python
#Changing the color
plt.scatter(x_values , y_values , c = 'red' , s = 43)
plt.show()
```

```python
plt.scatter(x_values , y_values , c = ( 0.91, 0.97,0.97) , s = 40)
plt.show()

#values closer to 0 will produce dark colors

```

```python
#Using ColorMap
x_values = [0] *20
y_values = list(range(20))
print(x_values)
print(y_values)
plt.scatter(x_values , y_values , c = x_values , cmap = plt.cm.Reds , s = 40)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
pts = np.random.randn(20)
plt.plot(pts, color = 'green', marker = 'd', markersize = 10)
plt.show()
```
